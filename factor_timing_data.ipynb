```python
from sklearn.decomposition import PCA
my_model = PCA(n_components=5)
my_model.fit_transform(df_T)

print (my_model.explained_variance_ratio_.cumsum())
y = np.matmul(df_T,np.transpose(my_model.components_))
# y.plot()
T_PC1 = y[0]
T_PC1.plot()
df_PCA = pd.DataFrame()
df_PCA = pd.concat([df_PCA,T_PC1],axis=1)
df_PCA
```
